```python
ㅁ
import seaborn as sns
import matplotlib.pyplot as plt

df_corr = df.drop('userID', axis=1)  # 'userID' 열 제거
df_corr = df_corr.drop('groupID', axis=1)  # 'userID' 열 제거

corr_matrix = df_corr.corr()  # 상관계수 행렬 계산

plt.figure(figsize=(10, 8))  # 그래프 크기 조정
sns.set(style='white')  # 그래프 스타일 설정

heatmap = sns.heatmap(
    corr_matrix,  # 상관계수 행렬 데이터
    annot=True,  # 각 셀에 상관계수 값 표시
    fmt='.2f',  # 소수점 2자리까지 표시
    cmap='coolwarm',  # 컬러맵 설정
    vmin=0, vmax=1,  # 컬러바 범위 설정
    linewidths=1,  # 셀 테두리 두께 설정
    cbar_kws={"shrink": .8},  # 컬러바 크기 조정
    square=True  # 정사각형 모양으로 그리드 설정
)

# 그래프 제목 설정
plt.title('Correlation Matrix', fontsize=18)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# 그래프 출력
plt.show()
```

```python

corr_matrix = df_corr.corr()

# 상관 행렬 히트맵 시각화
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
grouped_df = df.groupby('groupID').mean().reset_index()
grouped_df_melted = grouped_df.melt(id_vars='Group', var_name='Variable', value_name='Average Value')

plt.figure(figsize=(10, 6))
sns.barplot(data=grouped_df_melted, x='Group', y='Average Value', hue='Variable', palette='viridis')

# 그래프 제목과 축 레이블 설정
plt.title('Average Values by Group', fontsize=16)
plt.xlabel('Group', fontsize=14)
plt.ylabel('Average Value', fontsize=14)

# 범례 위치 조정
plt.legend(title='Variable', bbox_to_anchor=(1.05, 1), loc='upper left')

# 그래프 보여주기
plt.show()
```
